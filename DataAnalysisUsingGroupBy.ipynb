Understanding use of GroupBy in Pandas

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
df = pd.read_csv('googleplaystore.csv')
impute = SimpleImputer(missing_values = np.nan, strategy = 'mean')
impute.fit(df.iloc[: , 2:3].values)                                                                 
df.iloc[: , 2:3] = impute.transform(df.iloc[: , 2:3].values)

df = df.dropna()
df.head()
```

Q1. Name the apps that are in ART_AND_DESIGN with rating more than 4.5 in ascending order with respect to Ratings?

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
df_pr = df_pr[df_pr['Rating'] > 4.5]

df_pr.sort_values(by = 'Rating', ascending = True) 
```

Q2. Which category is having maximum average rating?

```python
df.groupby('Category').mean()['Rating'].sort_values(ascending = False)
```

Q3. How many free apps are there in each category in Descending order?

```python
df_pr = df[df['Type'] == 'Free']
df_pr.groupby('Category').count()['Type'].sort_values(ascending = False)
```
