HANDLING NULL VALUES - NUMERIC

1. You can handle by using dropna()

2. Or you can fill the rows or columns, here we wil see how to fill the null values using sklearn

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("googleplaystore.csv")
```

```python
df
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')                   #Initialising the imputer
```

```python
df.iloc[:,2:3].values                                  ##filling the values
```

```python
imputer.fit(df.iloc[:,2:3].values)

x = imputer.transform(df.iloc[:,2:3].values)            ## Transforming values using mean strategy
```

```python
x
```

```python
df.iloc[:,2:3] = x                     ##Assigning all the random values back to dataframe
```

```python
df
```

For most frequent we can replace all null values in one go meaning every null value would be replaced by the most frequent value in any numeri or categoric column

```python
ds = pd.read_csv("googleplaystore.csv")
```

```python
ds
```

```python
ds.isnull().sum()
```

```python
imputer2 = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputer.fit(ds.iloc[:,:].values)                                                                  ##method to replace all null values in both numeric and categoric form
ds.iloc[:,:] = imputer.transform(ds.iloc[:,:].values)
```

```python
ds
```

```python
ds.isnull().sum()               ##All null values are replaced with the most frquent one
```
