DATA ANALYSIS USING CONDITIONS 

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')
impute = SimpleImputer(missing_values = np.nan, strategy = 'mean')
impute.fit(df.iloc[: , 2:3].values)                                                                 
df.iloc[: , 2:3] = impute.transform(df.iloc[: , 2:3].values)

df = df.dropna()
```

```python
df.head()
```

Q1. How many free apps are there in ART_AND_DESIGN

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
print(len(df_pr[df_pr['Type'] == 'Free']))
```

Q2. How many apps are there in ART_AND_DESIGN with rating more than 4.5?

```python
df_pr = df[df['Rating'] > 4.5]
print(len(df_pr[df_pr['Category'] == 'ART_AND_DESIGN']))
```

Q3. How many apps are there in ART_AND_DESIGN with rating more than 4.5 and Free?

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
df_pr = df_pr[df_pr['Rating'] > 4.5]
print(len(df_pr[df_pr['Type'] == 'Free']))
```
